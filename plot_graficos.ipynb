```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
def plotar(arquivo, label):
    df = pd.read_csv(arquivo, delim_whitespace=True, header=None, names=['Episódio','Recompensa média'])
    plt.plot(df['Episódio'], df['Recompensa média'], label=label)

plt.figure(figsize=(15,7))
plt.xlabel('Episódio', fontsize=16)
plt.ylabel('Recompensa média', fontsize=16)
plt.grid(True)
plt.title('Q-Learning - input0', fontsize=20)

plotar('rewards/Entries/input0.txt_0.1_0.9_0.1_20000.txt', 'α=0.1, λ=0.9, ε=0.1')
plotar('rewards/Entries/input0.txt_0.1_0.9_0.5_20000.txt', 'α=0.1, λ=0.9, ε=0.5')
plotar('rewards/Entries/input0.txt_0.1_0.9_0.9_20000.txt', 'α=0.1, λ=0.9, ε=0.9')

plotar('rewards/Entries/input0.txt_0.5_0.9_0.1_20000.txt', 'α=0.5, λ=0.9, ε=0.1')
plotar('rewards/Entries/input0.txt_0.5_0.9_0.5_20000.txt', 'α=0.5, λ=0.9, ε=0.5')
plotar('rewards/Entries/input0.txt_0.5_0.9_0.9_20000.txt', 'α=0.5, λ=0.9, ε=0.9')

plotar('rewards/Entries/input0.txt_0.9_0.9_0.1_20000.txt', 'α=0.9, λ=0.9, ε=0.1')
plotar('rewards/Entries/input0.txt_0.9_0.9_0.5_20000.txt', 'α=0.9, λ=0.9, ε=0.5')
plotar('rewards/Entries/input0.txt_0.9_0.9_0.9_20000.txt', 'α=0.9, λ=0.9, ε=0.9')

plotar('rewards/Entries/input0.txt_0.9_0.1_0.5_20000.txt', 'α=0.9, λ=0.1, ε=0.9')

plt.legend()
plt.savefig('grafico1.png',format='png', bbox_inches='tight')
plt.show()


plt.figure(figsize=(15,7))
plt.xlabel('Episódio', fontsize=16)
plt.ylabel('Recompensa média', fontsize=16)
plt.grid(True)
plt.title('Q-Learning - input4', fontsize=20)

plotar('rewards/Entries/input4.txt_0.1_0.9_0.1_20000.txt', 'α=0.1, λ=0.9, ε=0.1')
plotar('rewards/Entries/input4.txt_0.1_0.9_0.5_20000.txt', 'α=0.1, λ=0.9, ε=0.5')
plotar('rewards/Entries/input4.txt_0.1_0.9_0.9_20000.txt', 'α=0.1, λ=0.9, ε=0.9')

plotar('rewards/Entries/input4.txt_0.5_0.9_0.1_20000.txt', 'α=0.5, λ=0.9, ε=0.1')
plotar('rewards/Entries/input4.txt_0.5_0.9_0.5_20000.txt', 'α=0.5, λ=0.9, ε=0.5')
plotar('rewards/Entries/input4.txt_0.5_0.9_0.9_20000.txt', 'α=0.5, λ=0.9, ε=0.9')

plotar('rewards/Entries/input4.txt_0.9_0.9_0.1_20000.txt', 'α=0.9, λ=0.9, ε=0.1')
plotar('rewards/Entries/input4.txt_0.9_0.9_0.5_20000.txt', 'α=0.9, λ=0.9, ε=0.5')
plotar('rewards/Entries/input4.txt_0.9_0.9_0.9_20000.txt', 'α=0.9, λ=0.9, ε=0.9')

plotar('rewards/Entries/input4.txt_0.9_0.1_0.5_20000.txt', 'α=0.9, λ=0.1, ε=0.9')
plt.legend()
plt.savefig('grafico2.png',format='png', bbox_inches='tight')
plt.show()


plt.figure(figsize=(15,7))
plt.xlabel('Episódio', fontsize=16)
plt.ylabel('Recompensa média', fontsize=16)
plt.grid(True)
plt.title('Q-Learning - input6', fontsize=20)

plotar('rewards/Entries/input6.txt_0.1_0.9_0.1_30000.txt', 'α=0.1, λ=0.9, ε=0.1')
plotar('rewards/Entries/input6.txt_0.1_0.9_0.5_30000.txt', 'α=0.1, λ=0.9, ε=0.5')
plotar('rewards/Entries/input6.txt_0.1_0.9_0.9_30000.txt', 'α=0.1, λ=0.9, ε=0.9')

plotar('rewards/Entries/input6.txt_0.5_0.9_0.1_30000.txt', 'α=0.5, λ=0.9, ε=0.1')
plotar('rewards/Entries/input6.txt_0.5_0.9_0.5_30000.txt', 'α=0.5, λ=0.9, ε=0.5')
plotar('rewards/Entries/input6.txt_0.5_0.9_0.9_30000.txt', 'α=0.5, λ=0.9, ε=0.9')

plotar('rewards/Entries/input6.txt_0.9_0.9_0.1_30000.txt', 'α=0.9, λ=0.9, ε=0.1')
plotar('rewards/Entries/input6.txt_0.9_0.9_0.5_30000.txt', 'α=0.9, λ=0.9, ε=0.5')
plotar('rewards/Entries/input6.txt_0.9_0.9_0.9_30000.txt', 'α=0.9, λ=0.9, ε=0.9')

plotar('rewards/Entries/input6.txt_0.9_0.1_0.5_30000.txt', 'α=0.9, λ=0.1, ε=0.9')
plt.legend()
plt.savefig('grafico3.png',format='png', bbox_inches='tight')
plt.show()


# plt.figure(figsize=(15,7))
# plt.xlabel('Episódio', fontsize=16)
# plt.ylabel('Recompensa média', fontsize=16)
# plt.grid(True)
# plt.title('Q-Learning - Caso do labirinto 11x20', fontsize=20)

# plotar('rewards/pacmaze4.txt_0.1_0.1_30000_30000.txt', 'α=0.1, ε=0.1')
# plotar('rewards/pacmaze4.txt_0.1_0.5_30000_30000.txt', 'α=0.1, ε=0.5')
# plotar('rewards/pacmaze4.txt_0.1_0.9_30000_30000.txt', 'α=0.1, ε=0.9')
# plotar('rewards/pacmaze4.txt_0.5_0.1_30000_30000.txt', 'α=0.5, ε=0.1')
# plotar('rewards/pacmaze4.txt_0.5_0.5_30000_30000.txt', 'α=0.5, ε=0.5')
# plotar('rewards/pacmaze4.txt_0.5_0.9_30000_30000.txt', 'α=0.5, ε=0.9')
# plotar('rewards/pacmaze4.txt_0.9_0.1_30000_30000.txt', 'α=0.9, ε=0.1')
# plotar('rewards/pacmaze4.txt_0.9_0.5_30000_30000.txt', 'α=0.9, ε=0.5')
# plotar('rewards/pacmaze4.txt_0.9_0.9_30000_30000.txt', 'α=0.9, ε=0.9')
# plt.legend()
# plt.savefig('doc/grafico4.pdf',format='pdf', bbox_inches='tight')
# plt.show()
```
